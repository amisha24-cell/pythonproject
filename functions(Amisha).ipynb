### Write an adapter function that takes a list of operations and executes add, sub, mul, div based on input. 

### Example: [ {"op": "add", "a": 10, "b": 20}, {"op": "mul", "a": 4, "b": 5} ]

```python
"""adapter function in python is a design pattern that allows incompatible interfaces to work together. 
It acts as a bridge between two incompatible interfaces by converting the interface of one class into an interface expected
"""
def sum(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    return a / b
def adapter(func, a, b):
    return func(a, b)
result1 = adapter(sum, 5, 3)          # Calls sum(5, 3)
result2 = adapter(subtract, 5, 3)     # Calls subtract(5, 3)
result3 = adapter(multiply, 5, 3)     # Calls multiply(5, 3)
result4 = adapter(divide, 5, 3)       # Calls divide(5, 3)

print(result1)
print(result2)
print(result3)
print(result4)
```

### Write a function process_students(data) that prints total marks and grade for each student.
- Example: [ {"name":"A", "marks":[80,90,70]}, {"name":"B", "marks":[60,75,65]} ] 
- Expected Output: Name: A | Total: 240 | Grade: A Name: B | Total: 200 | Grade: B

```python
def process_students(data):
    for student in data:
        name = student['name']
        marks = student['marks']
        total_marks = sum(marks)

        if total_marks > 200:
            print("student passed:", total_marks)
        else:
            print("student failed:", total_marks)


students_data = [
    {'name': 'A', 'marks': [80, 90, 70]},
    {'name': 'B', 'marks': [60, 75, 65]},
]

process_students(students_data)
```

### Create a function apply_discounts(products) that updates price after discount. 
- Example: [ {"name":"Laptop", "price":50000, "discount":10}, {"name":"Phone", "price":20000, "discount":5} ] 
- Expected Output: Laptop final: 45000 Phone final: 19000

```python
def apply_discount(products):
    for items in products:
        pname = items['name']
        price = items['price']
        discount = items['discount']

        updated_price = price - (price * discount / 100)
        print(f"Product: {pname}, Original Price: {price}, Discounted Price: {updated_price}")
products_list = [
    {'name': 'laptop', 'price': 50000, 'discount': 10},
    {'name': 'Phone', 'price': 20000, 'discount': 5}]
apply_discount(products_list)
```

### Write a function highest_salary(emp_list) that returns the employee with the highest salary.

```python
def highest_salary(emp_list):
    return max(emp_list, key=lambda emp: emp['salary'])
highest_salary(employees_list)
```

### Write a function average_salary(emp_list) that returns the average salary.

```python
def average_salary(emp_list):
    total_salary = sum(emp['salary'] for emp in emp_list)
    average_salary = total_salary / len(emp_list)
    return average_salary
average_salary(employees_list)
```

### Write a function salary_hike(emp_list, percent) that increases each salary by given percent.

```python
def salary_hike(emp_list, percent):
    for emp in emp_list:
        emp['salary'] += emp['salary'] * (percent / 100)
    return emp_list
salary_hike(employees_list, 10)
```

### Write out_of_stock(items) that returns all products with stock < 20.

```python
def out_of_stock(items):
    low_stock_items = []

    for item in items:
        if item['stock'] < 20:
            low_stock_items.append(item)

    return low_stock_items
products = [
    {"name": "Laptop", "stock": 15},
    {"name": "Mouse", "stock": 50},
    {"name": "Keyboard", "stock": 10},
    {"name": "Monitor", "stock": 5}
]

result = out_of_stock(products)
print(result)
```

### Write search_product(items, name) → return matching product(s).

```python
def search_product(item,name):
    for product in item:
        if product['name'] == name:
            return product
    return "Product not found"
item_list = [
    {'name': 'laptop', 'price': 50000},
    {'name': 'Phone', 'price': 20000},
    {'name': 'Tablet', 'price': 15000}]
search_product(item_list, 'Phone')
```
